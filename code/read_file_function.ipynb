```python
def gen_dct(fname):
    
    global dct
    dct = {}
    
    fh = open(fname)
    flines = fh.readlines()
    
    for i in range(len(flines)):
        dct["key%s" % i] = flines[i]
        dct["key%s" % i] = dct["key%s" % i].split()
        dct["key%s" % i][0] = dct["key%s" % i][0][8:10]
        
    print(dct)

gen_dct("userlog.log")
```
